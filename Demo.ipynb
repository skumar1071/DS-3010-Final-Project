```python
# Load test data

import pandas as pd

df_test = pd.read_csv('listings_test_20.csv')
```

```python
import random

test_file_path = 'listings_test_20.csv'
original_file_path = 'Data/listingsunzipped.csv'
RANDOM_SEED = 45

feature_columns = [
    'neighbourhood_cleansed',
    'property_type',
    'room_type',
    'accommodates',
    'bedrooms',
    'bathrooms',
    'n_amenities',
    'host_is_superhost',
    'minimum_nights',
    'availability_365'
]

df_test = pd.read_csv(test_file_path, low_memory=False)
df_original = pd.read_csv(original_file_path, low_memory=False)

df_test['id'] = df_test['id'].astype(str)
df_original['id'] = df_original['id'].astype(str)

random_listing = df_test.sample(n=1, random_state=RANDOM_SEED).copy()
target_listing_id = random_listing['id'].iloc[0]

listing_info_test = df_test[df_test['id'] == target_listing_id]
listing_info_original = df_original[df_original['id'] == target_listing_id]

features_to_select = [col for col in feature_columns if col in listing_info_test.columns]
extracted_features = listing_info_test[features_to_select].transpose()

extracted_features.columns = ['Feature Value']

airbnb_url = listing_info_original['listing_url'].iloc[0]

print(f"Selected Attributes for Random Listing from Test Data Set:")
print(f"Airbnb URL: {airbnb_url}")
print("-" * 50)
print(extracted_features.to_string())
print("-" * 50)
```
